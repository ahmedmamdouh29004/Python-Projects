```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
df = pd.read_csv("user-demographics.csv")
```

```python
df['Town/City'].str.strip()
df['Country'].str.strip()
```

```python
df = df[~df['Town/City'].isin(['64162', '(not set)'])]
```

```python
df.hist(figsize = (20,10),bins = 10)
plt.tight_layout()
plt.show()
```

```python
top_20_countries = df.sort_values(by = 'Country',ascending = False).head(20)
sns.barplot(y = 'Country' ,x = 'New users',data = top_20_countries,order = ['Yemen','Zimbabwe','Zambia','Western Sahara'])
plt.tight_layout()
plt.show()
```

```python
top_20_countries = df.sort_values(by = 'Users',ascending = False).head(20)
sns.barplot(y = 'Country' ,x = 'Users',data = top_20_countries,palette = 'crest',errorbar=None)
plt.tight_layout()
plt.show()
```

```python
top_cities = df.sort_values(by ='Users',ascending = False).head(20) 
sns.barplot(y =top_cities['Town/City'] , x = top_cities['Users'], data = top_cities,palette='viridis')
plt.xlabel("Number of Users")
plt.ylabel("Town/City")
plt.title("Top 20 City by Number of Users")
plt.figure(figsize = (20,10))
plt.show()
 
```

```python
top_20_cities = df.sort_values(by ='New users',ascending = False).head(20) 
sns.barplot(y ='Town/City' , x = 'New users', data = top_20_cities,palette = 'dark:salmon_r')
plt.xlabel("Number of New Users")
plt.ylabel("Town/City")
plt.title("Top 20 Cities by Number of New Users")
plt.figure(figsize = (20,10))
plt.show()
```

```python
sns.heatmap(data  =df.corr(numeric_only = True),annot=True, cmap='coolwarm', fmt=".2f" )
plt.figure(figsize=(20, 10))
plt.show()
```

```python
# top_engagement = df.sort_values(by = 'Engagement rate',ascending = False).head(20)
sns.barplot(y =top_20_countries['Town/City'],x = top_20_countries['Engagement rate'],palette = 'viridis')
plt.xlabel('Engagement Rate')
plt.ylabel('Town/City')
plt.title('Engagement Rate in Top 20 Cities by Number of Users')
plt.show()
```

```python
sns.relplot(x = df['Engagement rate'],y = df['Users'],kind = 'line')
```

```python
# top_Engaged_sessions = df.sort_values(by  = 'Engaged sessions per user',ascending = False).head(20)
plt.figure(figsize=(12, 8))
sns.barplot(x=top_20_countries['Engaged sessions per user'], y = top_20_countries['Town/City'],palette = 'viridis')
plt.xlabel('Engaged Sessions per User')
plt.ylabel('Town/City')
plt.title('Engaged Sessions per User in Top 20 Cities by Number of Users')
plt.show()
```

```python
# top_Average_Engagement_time = df.sort_values(by = 'Users',ascending = False).head(20)
plt.figure(figsize=(12, 8))
sns.barplot(y=top_20_countries['Town/City'], x=top_20_countries['Average engagement time'], palette='crest')
plt.xlabel('Average Engagement Time (seconds)')
plt.ylabel('Town/City')
plt.title('Average Engagement Time in Top 20 Cities by Number of Users')
plt.show()
```
